```python
weather = pd.read_csv("data/weather.csv")
```

```python
weather.head()
```

```python
weather_long = weather.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp"
)
weather_long.head()
```

```python
weather_tidy = weather_long.pivot_table(
    index=["id", "year", "month", "day"],
    columns="element",
    values="temp"
)
```

```python
weather_tidy.reset_index().head()
```

```python
weather_tidy = (weather
                .melt(
                    id_vars=["id", "year", "month", "element"],
                    var_name="day",
                    value_name="temp")
                .pivot_table(
                    index=["id", "year", "month", "day"],
                    columns="element",
                    values="temp")
                .reset_index()
)
```
